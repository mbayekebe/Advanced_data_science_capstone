## STEP 5 Train Test Split for Model Development

```python
#Splitting the dataset into the Training set and the Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

## STEP 6 Import Modules for Model Development

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

## STEP 7 Initialize Sequential Model

```python
#Initializing Neural Network
classifier = Sequential()
```

## STEP 8 Adding the layers one by one

```python
# first input layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu', input_dim = 36))
```

```python
# hidden layer

classifier.add(Dense(units = 3, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# output layer

classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'softmax'))
```

## STEP 9 Compiling the  Artificial Neural Network

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics =['accuracy'])
```

## STEP 10 Fitting, predicting for the Model

```python
# fitting the ann
classifier.fit(X_train, y_train, batch_size = 10, epochs = 25)
```

```python
# 
score = classifier.evaluate(X_test, y_test, verbose=0)
print('Test Loss:', score[0])
print('Test Accuracy:', score[1])
```

```python
# predicting and evaluating model
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

## STEP 11 Predicting test set result

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

## STEP 12 Evaluate model performance

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))
print(cm)
```
